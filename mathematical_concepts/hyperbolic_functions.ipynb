Hyperbolic functions are counterparts to trigonometric functions, but instead of being derived from circles, they are based on hyperbolas. These functions play a significant role in mathematics, physics, and engineering.

The hyperbolic sine $y = \sinh(x)$ and hyperbolic cosine $y = \cosh(x)$ share many properties with their trigonometric counterparts $y = \sin(x)$ and $y = \cos(x)$, respectively.

In the fields of machine learning and artificial intelligence, the hyperbolic tangent function $ \tanh(x)$ is commonly used as an activation function in neural networks due to its smooth and bounded output.


Hyperbolic Functions



$$
\cosh(x) = \frac{1}{2} (e^{x} + e^{-x})
$$

$$
\sinh(x) = \frac{1}{2} (e^{x} - e^{-x})
$$

$$
 \cosh(x) + \sinh(x) = e^{x}
$$

$$
 \cosh(x) - \sinh(x) = e^{-x}
$$


$$
\tanh(x) = \frac{\sinh(x)}{\cosh(x)} =  \frac{e^{x} - e^{-x}}{e^x + e^{-x}}
$$

$$
\cosh^2x - \sinh^2x = 1
$$

$$
\cosh^2x + \sinh^2x = 
$$
$$
(\frac{e^x + e^{-x}}{2})^2 + (\frac{e^x - e^{-x}}{2})^2 = \\
(\frac{e^{2x} + e^{-2x}+2}{4}) + (\frac{e^{2x} + e^{-2x}-2}{4}) = 
(\frac{2e^{2x} + 2e^{-2x}}{4}) = \\
\frac{e^{2x} + e^{-2x}}{2} = \cosh{2x}
$$

$$
\text{Derivative of} \sinh{x}: \text{  }
\text{    Let  } y= \sinh{x} = \frac{e^x - e^{-x}}{2}  \text{  } \text{    Then   }  \frac{dy}{dx} =  \frac{e^x + e^{-x}}{2} = \cosh{x}
$$

$$ 
\text{Derivative of} \cosh{x}: \text{  }
\text{    Let  } y= \cosh{x} = \frac{e^x + e^{-x}}{2}  \text{  } \text{    Then   }  \frac{dy}{dx} =  \frac{e^x - e^{-x}}{2} = \sinh{x}
$$


$$ 
\text{ Derivative of  } \sinh^{-1}x:  \text{  }

```python
def cosh(x):
    return (np.power(np.e,x) + np.power(np.e,-x)) / 2 

def sinh(x):
    return (np.power(np.e,x) - np.power(np.e,-x)) / 2

def tanh(x):
    return sinh(x)/cosh(x)
```

```python
n = [i * 0.001 for i in range(-2000,2000)]
assert {round( -sinh(x)**2 + cosh(x)**2 , 1) for x in n} == {1,}
```

```python
xpoints = np.array(n)
e_pow_x = np.array([np.power(np.e,x)  for x in n])
e_pow_minus_x = np.array([np.power(np.e,-x)  for x in n])
cosh_x = np.array([cosh(x) for x in n])
sinh_x = np.array([sinh(x) for x in n])

plt.plot(xpoints, e_pow_x, linestyle = 'dotted',label='e^x')
plt.plot(xpoints, e_pow_minus_x, linestyle = 'dotted',label='e^-x')
plt.plot(xpoints, cosh_x,label='cosh')
plt.plot(xpoints, sinh_x, label= 'sinh')
display_plot(plt)
```

```python
cosh_2x = np.array([cosh(x)**2+sinh(x)**2 for x in n])

plt.plot(xpoints, cosh_x,label='cosh')
plt.plot(xpoints, cosh_2x,label='cos2h')
display_plot(plt)
```

```python
tanh_x = np.array([tanh(x) for x in n])

plt.plot(xpoints, tanh_x,label='tanh')
display_plot(plt)
```

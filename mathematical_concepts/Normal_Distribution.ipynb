Applications: 

1. Used to model observational or measurement errors.
2. Serves as an approximation for the probability distribution of the mean of independent random variables.

The probability density function (PDF) of the normal distribution is given by:

$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{-\frac{(x-\mu)^2}{2\sigma^2}} $$
where:
1. $x$ is the random variable
2. $\mu$ is the mean
3. $\sigma^2$ is the variance

The cumulative distribution function (CDF) of the normal distribution is given by:

$$\int_{-\infty}^{x} \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{-\frac{(t-\mu)^2}{2\sigma^2}} \, dt $$

where:
1.  $x$ is the upper limit of integration
2.  $\mu$ is the mean
3. $\sigma^2$ is the variance

```python
x = np.linspace(-4, 4, 100)
y = scipy_stats.norm.pdf(x)

plt.plot(x, y)
```

```python
x = np.linspace(-4, 4, 100)
y = scipy_stats.norm.cdf(x)

plt.plot(x, y)
```

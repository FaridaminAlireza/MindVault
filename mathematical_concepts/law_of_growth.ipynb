<h3>Compound Interest Law of growth</h3>

Compound interest provides an excellent real-world context for understanding exponential growth. The core concept is that the value increases by a fixed percentage each period.
<br><br>
There are two main methods of calculating interest: simple interest and compound interest. In both cases, interest is applied as a fixed percentage of the principal—the initial amount of money invested or borrowed.

1. Simple Interest: The principal remains constant over time, and interest is calculated only on this original amount.
2. Compound Interest: Interest is added to the principal at the end of each period, and future interest is calculated on the updated total (principal + accumulated interest), leading to exponential growth.
<br>

Let P be the principal, and r be the rate percent per year.<br>
1. After 1 year, the added interest is $P * (r/100)$.
2. After 1 year, the total amount is $\quad (P + P * r/100) = P (1+ r/100)$. This is the principal for the new year.
3. After 2 years, the total amount is $\quad P(1 + r/100)^2$. 
4. After 3 years, the total amount is $\quad P(1+r/100)^3$.
5. After 4 years, the total amount is $\quad P(1+r/100)^4$. 

Now, suppose the interest is added every six months instead of once a year. Then:<br>
1. After 0.5 year, the total amount is $\quad P(1+r/2*100)$. 
2. After 1 year, the total amount is $\quad P(1+r/2*100)^2$. 
3. After 2 years, the total amount is $\quad P(1+r/2*100)^4$. 
4. After t year, the total amount is $\quad P(1+r/2*100)^{2t}$.

If the interest is added per month:<br>
1. After 1 year, the total amount is $P(1 +r/12*100)^{12}$. <br>
2. After t years, the total amount is $P(1+r/12*100)^{12t}$. <br>

If the interest is added m times a year:
- After t years, the total amount is $P(1+ r/100.m)^{mt}$.

- In $P(1+ r/100.m)^{mt}$, let = $r/100m = 1/n$, so that $m = nr/100$.
- After t years, the total amount is $P(1 + 1/n)^{nrt/100} = P{((1 + 1/n)^n)}^{(rt)/100}$.

- Now, suppose that n becomes infinitely large—that is, interest is added at infinitely small intervals.<br>
In this case, the growth of the principal can be considered continuous, and the total amount after t years is given by the limit of:
<br><br>


Now consider the case where interest is compounded an infinite number of times per year. That is, $ n $ becomes infinitely large, and interest is added at infinitesimally small intervals.  
In this case, the growth of the principal can be regarded as continuous, and the total amount after $ t $ years approaches the limit:

$
A = P \left( \left(1 + \frac{1}{n} \right)^n \right)^{\frac{rt}{100}}
$<br>

To evaluate this expression, we must determine the limit: $ \lim_{n \to \infty} \left(1 + \frac{1}{n} \right)^n$<br>

As $ n$ increases without bound: $
\lim_{n \to \infty} \left(1 + \frac{1}{n} \right)^n = e $ <br>

Substituting this result into the expression for $A$, we obtain: $$ A = Pe^{\frac{rt}{100}} $$

Letting $ x = \frac{rt}{100} $, we can rewrite the formula as:

$$
A = Pe^x
$$

<h3>The Exponential Function</h3>

The function $ e^x $ is called an exponential function because the variable appears in the exponent.

<h3>Continues Growth</h3>

Under continuous compounding, the growth of the principal occurs smoothly over time, rather than through discrete increases at regular intervals.

This kind of continuous growth is a fundamental principle in natural and scientific processes. In many physical, chemical, electrical, and engineering contexts, the rate of change of a quantity is proportional to the quantity itself. In such cases, the mathematical models often involve exponential functions.


### Exponential Distribution

**Example:**  
Imagine a chemical reactor. Let $ v $ be the effluent volumetric flow rate, i.e., the volume that leaves the reactor over a time interval $[0, t]$ is $ vt $ (an equal volume enters the vessel at the other end). Let $ V $ be the volume of the reactor vessel.

Then, a fraction  
$$
\left( \frac{v}{V} \right) t
$$  
will have left the vessel during $[0, t]$, assuming $ t $ is not too large.  

Let the random variable $ T $ represent the residence time of a particle in the vessel. To compute the distribution of $ T $, divide the interval $[0, t]$ into $ n $ small intervals of equal length $ \frac{t}{n} $.  

Assuming perfect mixing (so that the particle's position is uniformly distributed over the volume), the probability that a particle leaves the vessel during any one of these intervals is:  
$$
p = \left( \frac{v}{V} \right) \frac{t}{n}
$$

If the particle’s behavior is independent across intervals, and “leaving the vessel” is considered a success, then $ T $ follows a geometric distribution with success probability $ p $. The probability that the particle is still in the vessel at time $ t $ is approximately:

$$
P(T > t) \approx (1 - p)^n = \left( 1 - \frac{vt}{Vn} \right)^n
$$

Taking the limit as $ n \to \infty $:

$$
P(T > t) = \lim_{n \to \infty} \left( 1 - \frac{vt}{Vn} \right)^n = e^{-\frac{v}{V}t}
$$

Hence, the cumulative distribution function (CDF) of $ T $ is:

$$
F_T(t) = 1 - e^{-\frac{v}{V}t}
$$

By differentiating, we obtain the probability density function (PDF):

$$
f_T(t) = \frac{d}{dt} \left(1 - e^{-\frac{v}{V}t} \right) = \frac{v}{V} e^{-\frac{v}{V}t}, \quad t \geq 0
$$

This is an example of an **exponential distribution** with rate parameter $ \lambda = \frac{v}{V} $.

---

#### Exponential Distribution (General Definition)

A continuous random variable has an **exponential distribution** with rate parameter $ \lambda $ if its probability density function is defined as:

$$
f(x) = 
\begin{cases}
0, & x < 0 \\
\lambda e^{-\lambda x}, & x \geq 0
\end{cases}
$$

This distribution is denoted by $ \text{Exp}(\lambda) $.  

The cumulative distribution function (CDF) for an $ \text{Exp}(\lambda) $ random variable is:

$$
F(x) = 1 - e^{-\lambda x}
$$



<h4>Memoryless property of exponential distribution</h4>

If X has an exponential distribution, then for all s,t>0, 
$$P(X > s+t | X > s) = P(X > t)$$
This follows from:
$$P(X> s+t | X >s) = \frac{P(X>s+t)}{P(X>s)} = \frac{e^{- \lambda (s+t)}}{e^{-\lambda s}} = e^{- \lambda t} = P(X>t)$$


```python
def exp_density(l,a):
    return l * math.pow(math.e, -1 * l * a)
def exp_dist(l, a):
    return  1 - math.pow(math.e, -1 * l * a)
```

```python
l = 0.25
p= 20
x_points = [i for i in range(p)]
exp_dens= [exp_density(l,i) for i in x_points]

plt.scatter(x_points, exp_dens,label='exponential density function f(x)')
plt.plot(x_points, exp_dens)
display_plot(plt)
```

```python
l = 0.25
p= 20
x_points = [i for i in range(p)]
exp_distribution = [exp_dist(l,i) for i in x_points]

plt.scatter(x_points, exp_distribution,label='exponential distribution function F(x)')
plt.plot(x_points, exp_distribution)
display_plot(plt)
```

#### Applications of the Exponential Distribution

The exponential distribution is widely used to model the time between successive events in a Poisson process. Common applications include:

- **Waiting times**, such as the time until a system failure  
- **Decay processes**, such as radioactive decay or chemical reactions  
- **Queueing systems**, such as the time between customer arrivals  

Let $ T $ denote the waiting time until an event occurs—such as a molecule escaping from a volume $ V $, or a system transitioning to a new state. If events occur continuously and independently at a constant rate $ \lambda = \frac{v}{V} $, then $ T $ follows an exponential distribution with parameter $ \lambda $.

The **cumulative distribution function (CDF)** is:

$$
P(T \leq t) = 1 - e^{-\lambda t}
$$

This gives the probability that the event has occurred by time $ t $.

The **complementary probability** (also called the survival function) is:

$$
P(T > t) = e^{-\lambda t}
$$

This gives the probability that the event has not yet occurred by time $ t $.

Such expressions naturally arise in processes involving random waiting times, where events occur independently and continuously at a constant average rate.

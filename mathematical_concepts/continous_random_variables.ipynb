<h3>Continues random variables</h3>
A random variable $X$ is said to be **continuous** if there exists a function $ f: \mathbb{R} \to \mathbb{R} $ such that for any real numbers $ a \leq b $:

$$
P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx
$$

The function $ f $ must satisfy the following conditions:

- $ f(x) \geq 0 $ for all $ x \in \mathbb{R} $
- $ \int_{-\infty}^{\infty} f(x) \, dx = 1 $

The function $ f $ is called the **probability density function** (PDF) of the random variable $ X $.


### Distribution Function

For a continuous random variable $X$, the probability that $ X $ lies between $ a $ and $ b $ is given by:

$$
P(a < X < b) = P(X \leq b) - P(X \leq a) = F(b) - F(a)
$$

There is a simple relationship between the **distribution function** $ F $ and the **probability density function** $ f $ of a continuous random variable. Using integral calculus, we obtain:

$$
F(b) = \int_{-\infty}^{b} f(x) \, dx
$$
$$ \text{ and }
f(x) = \frac{d}{dx} F(x) $$


### Uniform Distribution

A continuous uniform distribution occurs when a random variable $ X $ is equally likely to take any value within the interval $ [\alpha, \beta] $.

A continuous random variable follows a uniform distribution on the interval $ [\alpha, \beta] $ if its probability density function (PDF) is defined as:

$$
f(x) = 
\begin{cases} 
0, & \text{if } x \notin [\alpha, \beta] \\
\frac{1}{\beta - \alpha}, & \text{if } \alpha \leq x \leq \beta
\end{cases}
$$

This distribution is denoted as $ U(\alpha, \beta) $.


```python
def uniform_density(alpha, beta):
    return 1/(beta - alpha)
def uniform_distribution(alpha, beta, x):
    return sum(1/(beta - alpha) for _ in range(alpha, x)) 
```

```python
alpha = 0
beta = 20
x_points = [i for i in range(alpha,beta)]
uniform_dens= [uniform_density(alpha, beta) for i in range(alpha,beta)]

plt.scatter(x_points, uniform_dens,label='uniform density f(x)')
plt.plot(x_points, uniform_dens)
display_plot(plt)
```

```python
alpha = 0
beta = 20
x_points = [i for i in range(alpha,beta)]
uniform_dens= [uniform_distribution(alpha, beta,i) for i in range(alpha,beta)]

plt.scatter(x_points, uniform_dens,label='uniform density f(x)')
plt.plot(x_points, uniform_dens)
display_plot(plt)
```

**Example of a Uniform Distribution:**

Suppose a bus is scheduled to arrive at a random time between 10:00 AM and 10:30 AM. If you arrive at a randomly chosen moment during this 30-minute interval, the bus arrival time $ X $ can be modeled by a uniform distribution on the interval $[0, 30]$, where time is measured in minutes after 10:00 AM.

That is, $ X \sim U(0, 30) $, and the probability density function is:

$$
f(x) = \frac{1}{30}, \quad \text{for } 0 \leq x \leq 30
$$
